```python
from ultralytics import YOLO

DATA_YAML = "/content/drive/MyDrive/Datasets/lpchardb/data.yaml"

model = YOLO("yolo11n.pt")

model.train(
    data=DATA_YAML,
    epochs=80,
    imgsz=640,
    batch=16,
    workers=8
)
```
